```python
import requests
import json

# URL of the reel post
reel_url = "https://www.instagram.com/p/C683FOGyPXu/"

# Make a GET request to the post URL
try:
    response = requests.get(reel_url)
    response.raise_for_status()  # Raise an exception for bad status codes
except requests.exceptions.RequestException as e:
    print("Error:", e)
else:
    # Parse the HTML content to find the JSON data containing reel information
    start_string = '"shortcode":"'
    end_string = '","'
    start_index = response.text.find(start_string)
    end_index = response.text.find(end_string, start_index + len(start_string))
    if start_index != -1 and end_index != -1:
        shortcode = response.text[start_index + len(start_string):end_index]
        graphql_url = "https://www.instagram.com/graphql/query/?query_hash=477b65a610463740ccdb83135b2014db&variables="
        graphql_query = json.dumps({"shortcode": shortcode, "child_comment_count": 3, "fetch_comment_count": 40, "parent_comment_count": 24, "has_threaded_comments": True})
        graphql_response = requests.get(f"{graphql_url}{graphql_query}")
        if graphql_response.status_code == 200:
            graphql_data = graphql_response.json()
            reel_data = graphql_data.get("data", {}).get("shortcode_media", {})
            view_count = reel_data.get("video_view_count", "N/A")
            print("View Count:", view_count)
        else:
            print("Failed to fetch reel data.")
    else:
        print("Reel shortcode not found.")
```

```python
import instaloader
import time
import pandas as pd

# Create an instance of Instaloader
loader = instaloader.Instaloader()

# Define the Instagram profile you want to scrape
profile_name = "fam.india"

# List to hold extracted data
data = []

# Load the profile
try:
    profile = instaloader.Profile.from_username(loader.context, profile_name)
except instaloader.exceptions.InstaloaderException as e:
    print("Error loading profile:", e)
else:
    # Iterate over all posts in the profile
    for post in profile.get_posts():
        # Check if the post is a reel (video)
        if post.is_video:
            try:
                # Extract details for the reel
                views = post.video_view_count
                likes = post.likes
                comments = post.comments
                caption = post.caption
                post_url = f"https://www.instagram.com/p/{post.shortcode}/"

                # Append the extracted details to the data list
                data.append({
                    "URL": post_url,
                    "Views": views,
                    "Likes": likes,
                    "Comments": comments,
                    "Caption": caption
                })

                # Wait a bit to avoid hitting rate limits
                time.sleep(2)
            except instaloader.exceptions.InstaloaderException as e:
                print("Error loading post:", e)

# Introduce a delay to allow time for data collection
time.sleep(10)

# Create a DataFrame from the collected data
df = pd.DataFrame(data)

# Check if the DataFrame is empty
if df.empty:
    print("No reels found for the specified profile.")
else:
    # Print the DataFrame to verify the output
    print(df)

# Now, you can save the data from the DataFrame to an Excel file when desired.
```

```python
import instaloader
import time
import pandas as pd

# Create an instance of Instaloader
loader = instaloader.Instaloader()

# Define the Instagram profile you want to scrape
profile_name = "fam.india"

# List to hold extracted data
data = []

# Load the profile
try:
    profile = instaloader.Profile.from_username(loader.context, profile_name)
except instaloader.exceptions.InstaloaderException as e:
    print("Error loading profile:", e)
else:
    # Iterate over all posts in the profile
    for post in profile.get_posts():
        if post.is_video:
            try:
                # Extract datetime and link of the reel
                post_datetime = post.date_utc
                reel_link = f"https://www.instagram.com/p/{post.shortcode}/"

                # Append the extracted data to the data list
                data.append({
                    "Datetime": post_datetime,
                    "Reel Link": reel_link
               })

                # Wait a bit to avoid hitting rate limits
                time.sleep(2)
            except instaloader.exceptions.InstaloaderException as e:
                print("Error loading post:", e)

# Introduce a delay to allow time for data collection
time.sleep(10)

# Create a DataFrame from the collected data
df = pd.DataFrame(data)

# Check if the DataFrame is empty
if df.empty:
    print("No reels found for the specified profile.")
else:
    # Print the DataFrame to verify the output
    print(df)

# Now, you can save the data from the DataFrame to an Excel file when desired.
```

```python
df = pd.read_excel(r'D:\youvah\other accounts\famindia_merged_sheet.xlsx')
```

```python
def extract_hashtags(caption):
    # Ensure the input is a string
    caption = str(caption)
    return re.findall(r'#\w+', caption)
```

```python
df['hashtags'] = df['Caption'].apply(lambda x: extract_hashtags(x))
```

```python
df['hashtags'] = df['hashtags'].apply(lambda x: ' '.join(x))
```

```python
df.to_excel('famindia_final.xlsx', index=False)
```

```python
def count_hashtags(hashtags):
    return len(hashtags.split())
```
